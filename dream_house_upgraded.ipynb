```python
import math
import turtle
t = turtle.Turtle()

t.speed(0)

# Vẽ ngôi nhà
t.fillcolor("blue")
t.begin_fill()
t.forward(200)
t.left(90)
t.forward(100)
t.left(90)
t.forward(200)
t.left(90)
t.forward(100)
t.end_fill()

# Vẽ cửa
t.fillcolor("yellow")
t.begin_fill()
t.left(90)
t.forward(70)
t.left(90)
t.forward(60)
t.right(90)
t.forward(50)
t.right(90)
t.forward(60)
t.end_fill()

# Vẽ ống khói
t.right(90)
t.forward(5)
t.right(90)
t.penup()
t.forward(100)
t.pendown()
t.fillcolor("grey")
t.begin_fill()
t.forward(60)
t.left(90)
t.forward(20)
t.left(90)
t.forward(60)
t.end_fill()

# Vẽ mái nhà
t.fillcolor("red")
t.begin_fill()
t.left(90)
t.forward(45)
t.left(135)
t.forward(math.sqrt(70*70*2))
t.left(90)
t.forward(math.sqrt(70*70*2))
t.end_fill()

# Vẽ cây nhỏ
    # Vẽ tán thứ 1
t.penup()
t.left(45)
t.forward(50)
t.right(90)
t.forward(20)
t.pendown()
t.fillcolor("green")
t.begin_fill()
t.setheading(135)
t.forward(math.sqrt(20*20*2))
t.left(90)
t.forward(math.sqrt(20*20*2))
t.left(135)
t.forward(40)
t.end_fill()
    # Vẽ tán thứ 2
t.right(90)
t.penup()
t.forward(30)
t.left(90)
t.forward(10)
t.pendown()
t.fillcolor("green")
t.begin_fill()
t.setheading(135)
t.forward(math.sqrt(30*30*2))
t.left(90)
t.forward(math.sqrt(30*30*2))
t.left(135)
t.forward(60)
t.end_fill()
    # Vẽ tán thứ 3
t.right(90)
t.penup()
t.forward(40)
t.left(90)
t.forward(10)
t.pendown()
t.fillcolor("green")
t.begin_fill()
t.setheading(135)
t.forward(math.sqrt(40*40*2))
t.left(90)
t.forward(math.sqrt(40*40*2))
t.left(135)
t.forward(80)
t.end_fill()
    # Vẽ thân cây
t.backward(30)
t.right(90)
t.fillcolor("brown")
t.begin_fill()
t.forward(40)
t.right(90)
t.forward(20)
t.right(90)
t.forward(40)
t.end_fill()

# Vẽ cây to
t.penup()
t.goto(400,150)
t.pendown()

    # Vẽ tán thứ 1
t.fillcolor("green")
t.begin_fill()
t.setheading(135)
t.forward(math.sqrt(60*60*2))
t.left(90)
t.forward(math.sqrt(60*60*2))
t.left(135)
t.forward(120)
t.end_fill()
    # Vẽ tán thứ 2
t.right(90)
t.penup()
t.forward(80)
t.left(90)
t.forward(20)
t.pendown()
t.fillcolor("green")
t.begin_fill()
t.setheading(135)
t.forward(math.sqrt(80*80*2))
t.left(90)
t.forward(math.sqrt(80*80*2))
t.left(135)
t.forward(160)
t.end_fill()
    # Vẽ tán thứ 3
t.right(90)
t.penup()
t.forward(100)
t.left(90)
t.forward(20)
t.pendown()
t.fillcolor("green")
t.begin_fill()
t.setheading(135)
t.forward(math.sqrt(100*100*2))
t.left(90)
t.forward(math.sqrt(100*100*2))
t.left(135)
t.forward(200)
t.end_fill()
    # Vẽ thân cây
t.backward(80)
t.right(90)
t.fillcolor("brown")
t.begin_fill()
t.forward(80)
t.right(90)
t.forward(40)
t.right(90)
t.forward(80)
t.end_fill()

# kết thúc vẽ
turtle.done()
```
